```python
import json
import os

from PIL import Image,ImageDraw
from tqdm import tqdm
from glob import glob
```

```python
def check_GT(txt_path, bmp_path):
    for name in tqdm(glob(os.path.join(txt_path, '*.txt'))):
        basename = os.path.basename(name)
        with open (name) as f:
            lines = f.readlines()
            label = []
            points = []
            print(name)
        for line in lines:
            each_label = []
            for order in line.split():
                each_label.append(order)
            label.append(each_label)        
#       print(label)
        
        image = Image.open(os.path.join(bmp_path, '{}.bmp'.format(basename[:-4])))
        draw = ImageDraw.Draw(image)
        width, height = image.size
    
        for each in label:
#             print(each)
            x = float(each[1]) * width
            y = float(each[2]) * height
            w = float(each[3]) * width
            h = float(each[4]) * height
            points = [(x - w /2,y - h / 2), (x + w / 2, y - h / 2), (x + w /2,y + h / 2), (x - w /2,y + h / 2)]
            draw.polygon(points, outline=(255, 0, 0))
    
        display(image)
#       image.save(save_path)


            


    
```
