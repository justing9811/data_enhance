```python
def aspect_ratio(path):
    value = []
    
    for name in tqdm(glob(os.path.join(path, '*.txt'))):
        
        with open (name) as f:
            lines = f.readlines()
            label = []
            
        
        for line in lines:
            each_label = []
            for order in line.split():
                each_label.append(order)
            label.append(each_label)        
        
        for each in label:
#             print(each)
            w = float(each[3]) 
            h = float(each[4]) 
            ratio = w / h
            
            if ratio > 1:
                ratio = 2 - 1 / ratio
                
        value.append(ratio)
    
    sort = []
    value.sort()
    for v in value:
        sort.append([v, value.count(v)])
    
    sort.sort(key=lambda x:x[0], reverse=True)
    
    sort = dict(sort)
    print(sort)
    
    
#     (x0, y0) = sort[0]
#     for (x, y) in sort:
#         print(x,y)
#         if x0 == x:
#             sort.remove([x, y])
#         else:
#             x0 = x
#             y0 = y
#     print(sort)

        
    # 绘制直方图
#     list_x = []
#     list_y = []
#     plt.xlim(0,2)
#     for (x,y) in sort:
#         list_x.append(x)
#         list_y.append(y)
#         plt.plot(list_x, list_y, 'r', lw=2)
#     plt.show()
    
    #kde 
#     k = kde.KernelDensity(kernel='gaussian', bandwidth=0.01).fit(np.array(value)[:, np.newaxis])
#     list_y = np.exp(k.score_samples(np.array([v / 1000 for v in range(2000)])[:, np.newaxis]))
    
#     plt.plot(np.array([v / 1000 for v in range(2000)]), list_y, 'r', lw=2)
#     plt.show()
    
    

    
```
