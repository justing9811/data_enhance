```python
import os
import shutil  # Python文件复制相应模块
```

```python
# 读取txt文件中的每一行，并根据读取的名称从文件夹中复制出相关图片
# input_pic_path = "D:/所读论文/代码/cascade-rcnn(modify)/data/VOCdevkit2007_9563/VOC2007/JPEGImages"
input_txt_path = "D:/data/hy/VOC12_13/VOC2007/total_txt"
fopen = open('D:/data/hy/VOC12_13/VOC2007/ImageSets/Main/val.txt', 'r')
lines = fopen.readlines()
# out_pic = "D:/所读论文/代码/cascade-rcnn(modify)/data/VOCdevkit2007_9563/VOC2007/test_images"
#存放提取出来的xml
out_txt = "D:/data/hy/VOC12_13/VOC2007/val_txt"
for line in lines:
    line = line.strip('\n')  # 去掉列表中每一个元素的换行符
    # pic_file_name = line + ".bmp"
    txt_file_name = line + ".txt"
    # print(pic_file_name)
    print(txt_file_name)
    # pic_file_path = os.path.join(input_pic_path, pic_file_name)
    txt_file_path = os.path.join(input_txt_path, txt_file_name)
    # shutil.copy(pic_file_path, out_pic)
    shutil.move(txt_file_path, out_txt)
```
